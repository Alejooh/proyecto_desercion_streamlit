```python
import numpy as np

df["PROMEDIO_NUM"] = (
    df["PROMEDIO"]
    .astype(str)
    .str.replace(",", ".", regex=False)
)

df["PROMEDIO_NUM"] = pd.to_numeric(df["PROMEDIO_NUM"], errors="coerce")

df[["PROMEDIO", "PROMEDIO_NUM"]].head(10)
```

```python
import re

def periodo_sort_key(p):
    p = str(p).strip()
    # Captura "2024 - 2025" y el sufijo (CI/CII/ING1A...)
    m = re.match(r"(\d{4})\s*-\s*(\d{4})\s*(.*)", p)
    if not m:
        return (9999, 999, p)
    y1 = int(m.group(1))
    suf = m.group(3).strip()

    # Orden recomendado: CI primero, luego CII, y después otros al final
    order_map = {"CI": 1, "CII": 2}
    suf_order = order_map.get(suf, 9)

    return (y1, suf_order, suf)

# Crear columna de orden
df["PERIODO_KEY"] = df["PERIODO"].apply(periodo_sort_key)

# Ver los periodos ordenados
sorted(df["PERIODO"].unique(), key=periodo_sort_key)
```

```python
agg = (
    df.groupby(["ESTUDIANTE", "PERIODO", "PERIODO_KEY"])
      .agg(
          facultad=("FACULTAD", "first"),
          carrera=("CARRERA", "first"),
          nivel=("NIVEL", "max"),
          materias_cursadas=("COD_MATERIA", "count"),
          promedio_periodo=("PROMEDIO_NUM", "mean"),
          asistencia_prom=("ASISTENCIA", "mean"),
          reprobadas=("ESTADO", lambda s: (s == "REPROBADA").sum()),
          aprobadas=("ESTADO", lambda s: (s == "APROBADA").sum()),
          max_no_vez=("NO. VEZ", "max"),
          repitencia_prom=("NO. VEZ", "mean"),
      )
      .reset_index()
)

agg["tasa_reprobacion"] = agg["reprobadas"] / agg["materias_cursadas"]
agg.head()
```

```python
import re

def tipo_periodo(p):
    p = str(p).strip()
    m = re.match(r"\d{4}\s*-\s*\d{4}\s*(.*)", p)
    suf = m.group(1).strip() if m else ""
    if suf in ["CI", "CII"]:
        return "REGULAR"
    else:
        return "ESPECIAL"

df["TIPO_PERIODO"] = df["PERIODO"].apply(tipo_periodo)
df["TIPO_PERIODO"].value_counts()
```

```python
next_map = {periods_reg[i]: periods_reg[i+1] for i in range(len(periods_reg)-1)}
next_map
```

```python
# Conjunto de pares (ESTUDIANTE, PERIODO) existentes
pairs = set(zip(agg_reg["ESTUDIANTE"], agg_reg["PERIODO"]))

def desercion_next(row):
    p = row["PERIODO"]
    if p not in next_map:
        return None  # último período no se etiqueta
    p_next = next_map[p]
    return 0 if (row["ESTUDIANTE"], p_next) in pairs else 1

agg_reg["DESERCION_NEXT"] = agg_reg.apply(desercion_next, axis=1)
agg_reg["DESERCION_NEXT"].value_counts(dropna=False)
```

```python
# data_ml: solo filas donde existe etiqueta DESERCION_NEXT
data_ml = agg_reg.dropna(subset=["DESERCION_NEXT"]).copy()
data_ml["DESERCION_NEXT"] = data_ml["DESERCION_NEXT"].astype(int)

# Split temporal: último PERIODO como test
periods_ml = sorted(data_ml["PERIODO"].unique(), key=periodo_sort_key)
last_period_ml = periods_ml[-1]
print("Test period:", last_period_ml)

train_df = data_ml[data_ml["PERIODO"] != last_period_ml].copy()
test_df  = data_ml[data_ml["PERIODO"] == last_period_ml].copy()

print("Train target:\n", train_df["DESERCION_NEXT"].value_counts())
print("Test target:\n",  test_df["DESERCION_NEXT"].value_counts())

y_train = train_df["DESERCION_NEXT"]
y_test  = test_df["DESERCION_NEXT"]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

num_features = [
    "nivel", "materias_cursadas", "promedio_periodo", "asistencia_prom",
    "tasa_reprobacion", "max_no_vez", "repitencia_prom"
]

X_train = train_df[num_features]
X_test  = test_df[num_features]

y_train = train_df["DESERCION_NEXT"]
y_test  = test_df["DESERCION_NEXT"]

preprocess = ColumnTransformer(
    transformers=[("num", "passthrough", num_features)],
    remainder="drop"
)

logit_model = Pipeline(steps=[
    ("prep", preprocess),
    ("clf", LogisticRegression(max_iter=2000, class_weight="balanced"))
])

logit_model.fit(X_train, y_train)
y_pred = logit_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, digits=3))
```
